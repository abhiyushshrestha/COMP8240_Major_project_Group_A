```
import wikipedia
import urllib.request
import numpy as np

Celebrity = wikipedia.page("list_of_facial_images_of_Celebrities").links

ext = (".jpg",".png")
print(len(Celebrity))
for Celebrity_specific in Celebrity:
    Celeb = wikipedia.page(Celebrity_specific).links
    np.random.shuffle(Celeb)
    for page in Celeb[:round(len(Celeb)/10)]:
      try:
        wikipage = wikipedia.page(page)
        print("Page Title: %s" % wikipage.title)
        print ("Page URL: %s" % wikipage.url)
        print (" - Main Image: %s" % wikipage.images[1])
        for count,image in enumerate(wikipage.images):
          if image.endswith(ext):
            urllib.request.urlretrieve(image,page+str(count)+".jpg")
        print ("")
      except:
        pass 

```

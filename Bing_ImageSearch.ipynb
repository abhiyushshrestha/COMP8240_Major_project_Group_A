```python
from azure.cognitiveservices.search.imagesearch import ImageSearchClient
from msrest.authentication import CognitiveServicesCredentials
```

```python
subscription_key = "d5775aed84e246b99f17713afbf57b99"
subscription_endpoint = "https://imagesearchtestp.cognitiveservices.azure.com/"
search_term = "Justin Bieber"
```

```python
client = ImageSearchClient(endpoint=subscription_endpoint, credentials=CognitiveServicesCredentials(subscription_key))
image_results = client.images.search(query=search_term)
```

```python
if image_results.value:
    print("Total number of images returned: {}".format(len(image_results.value)))
    for image in (image_results.value):
        print("Image thumbnail url: {}".format(image.thumbnail_url))
else:
    print("No image results returned!")
```

```python
search_terms = ["Justin Bieber", "Selena Gomez", "Gordon Ramsay", "Jennifer Lopez", "Tom Cruise", "Mr Bean", "Ed Sheeren", "Elon Musk", "Barack Obama", "Christiano Ronaldo"]

for search_term in search_terms:
    image_results = client.images.search(query=search_term)
    if image_results.value:
        print("Total number of images returned: {}".format(len(image_results.value)))
        for index, image in enumerate(image_results.value):
            image_url = image.thumbnail_url
            print("Image thumbnail url: {}".format(image_url))
            filename = search_term.split(" ")
            filename = "_".join(filename)
            urllib.request.urlretrieve(image_url, filename + "_" + str(index) + ".jpg")
    else:
        print("No image results returned!")
```
